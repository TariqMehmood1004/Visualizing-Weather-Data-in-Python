```python
from mpl_toolkits.basemap import Basemap
from matplotlib.offsetbox import AnnotationBbox, OffsetImage
import matplotlib.image as mpimg
import urllib
from io import BytesIO
from PIL import Image


import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import matplotlib.patheffects as path_effects
import urllib.request
import matplotlib.image as mpimg
from io import BytesIO
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
import requests

# Function to Get Current Location (Country & City) from IP
def get_current_location():
    try:
        response = requests.get("http://ip-api.com/json/").json()
        return response["country"], response["city"]
    except:
        return None, None  # Fallback if API fails

# Function to get all cities of a country
def get_cities_weather(country):
    cities = []
    API_URL = f"http://api.weatherapi.com/v1/search.json?key={API_KEY}&q={country}"
    response = requests.get(API_URL)

    if response.status_code == 200:
        cities_data = response.json()
        for city in cities_data:
            cities.append({
                "name": city["name"],
                "lat": city["lat"],
                "lon": city["lon"]
            })
    return cities
    
# Function to fetch weather data for a city
def get_weather_data(city):
    API_URL = f"http://api.weatherapi.com/v1/current.json?key={API_KEY}&q={city}&aqi=no"
    response = requests.get(API_URL)

    if response.status_code == 200:
        data = response.json()
        return {
            "name": data["location"]["name"],
            "lat": data["location"]["lat"],
            "lon": data["location"]["lon"],
            "temp_c": data["current"]["temp_c"],
            "icon": f"https:{data['current']['condition']['icon']}"
        }
    return None

#  Get Automatically Detected Country & City
country, city = get_current_location()
if not city:
    print("Could not detect location. Using default: Karachi, Pakistan")
    country, city = "Pakistan", "Karachi"

#  Fetch Weather Data
weather_data = get_weather_data(city)
if not weather_data:
    print("Weather data not found!")
else:
    print(f"Current temperature in {weather_data['name']}: {weather_data['temp_c']}C")
```

```python
# Get all cities' weather data for a country
#  Automatically Detect Country
country = get_current_location()

#  Fetch Weather Data for All Cities
cities = get_cities_weather(country)
weather_data = [get_weather_data(city) for city in cities if get_weather_data(city)]

# Create Map Figure
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 8), facecolor="white")

# Background Maps with Enhanced Colors
m1 = Basemap(projection='mill', llcrnrlon=60, llcrnrlat=20, urcrnrlon=80, urcrnrlat=40, ax=axes[0])
m1.drawcoastlines(linewidth=1.2, color="black")
m1.drawcountries(linewidth=1.2, color="black")
m1.fillcontinents(color="tan", lake_color="deepskyblue")
m1.drawmapboundary(fill_color="deepskyblue")

m2 = Basemap(projection='mill', llcrnrlon=60, llcrnrlat=20, urcrnrlon=80, urcrnrlat=40, ax=axes[1])
m2.drawcoastlines(linewidth=1.2, color="black")
m2.drawcountries(linewidth=1.2, color="black")
m2.fillcontinents(color="tan", lake_color="deepskyblue")
m2.drawmapboundary(fill_color="deepskyblue")

# Weather Icons and Temperatures on the Maps
for city in weather_data:
    lat, lon, temp, icon_url = city["lat"], city["lon"], city["temp_c"], city["icon"]

    try:
        # Fetch weather icon
        icon_data = urllib.request.urlopen(icon_url).read()
        icon_img = mpimg.imread(BytesIO(icon_data))
        imagebox = OffsetImage(icon_img, zoom=0.15)
        ab = AnnotationBbox(imagebox, m1(lon, lat), frameon=False)
        axes[0].add_artist(ab)
    except:
        pass

    # Convert coordinates for plotting
    x1, y1 = m2(lon, lat)  # Ensure proper coordinate conversion
    
    # Temperature color logic
    temp_color = "red" if temp > 25 else "orange" if temp > 15 else "blue"
    
    # Add temperature label
    bbox_props = dict(boxstyle="round,pad=0.4", fc=temp_color, ec="black", lw=2, alpha=0.8)
    axes[1].text(x1, y1, f"{temp}C", ha="center", va="center", fontsize=11, fontweight="bold", color="white",
                 bbox=bbox_props, path_effects=[path_effects.withStroke(linewidth=3, foreground="black")])
    
    # Add city name slightly above the temperature label
    axes[1].text(x1, y1 + 20000, city["name"], ha="center", va="bottom", fontsize=50, fontweight="bold", color="black")


plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05, wspace=0.3)
plt.show()
```
